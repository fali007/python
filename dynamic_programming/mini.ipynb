```python
from tkinter import *
import csv
import tkinter.messagebox

def get_the_details():
    details = [entry1.get(), entry2.get(), entry3.get(), entry4.get(), entry5.get()]
    data = open("details.csv", "a")
    with data:
        writer = csv.writer(data)
        writer.writerow(details)
    print("saved details")


def menu_bar():
    print("welcome nigga")

def check():
    if len(entry1.get()) == 0:
        tkinter.messagebox.showerror('ALERT', 'Please Fill all Features')
    elif len(entry2.get()) == 0:
        tkinter.messagebox.showerror('ALERT', 'Please Fill all Features')
    elif len(entry3.get()) == 0:
        tkinter.messagebox.showerror('ALERT', 'Please Fill all Features')
    elif len(entry4.get()) == 0:
        tkinter.messagebox.showerror('ALERT', 'Please Fill all Features')
    elif len(entry5.get()) == 0:
        tkinter.messagebox.showerror('ALERT', 'Please Fill all Features')
    else:
        get_the_details()


root = Tk()

#menu_bar
menu = Menu(root)
root.config(menu=menu)

subMenu = Menu(menu)
menu.add_cascade(label="Login", menu=subMenu)
subMenu.add_command(label="Student", command=menu_bar)
subMenu.add_command(label="Admin", command=menu_bar)
subMenu.add_command(label="Official", command=menu_bar)

subMenu.add_separator()
subMenu.add_command(label="exit", command=menu_bar)

editMenu = Menu(menu)
menu.add_cascade(label="Sign Up", menu=editMenu)
editMenu.add_command(label="Student", command=menu_bar)
editMenu.add_command(label="Admin", command=menu_bar)
editMenu.add_command(label="Official", command=menu_bar)

##toolbar
toolBar = Frame(root, bg="blue")
insertButton = Button(toolBar, text="Insert Image", command=menu_bar)
insertButton.pack(side=LEFT)
toolBar.pack(fill=X, side=TOP)


#status_bar
status = Label(root, text="there is nothing to do", bd=1, relief=SUNKEN, anchor=W)
status.pack(side=BOTTOM, fill=X)


#body
left_frame = Frame(root, width=500, height=100)
left_frame.pack(side=LEFT)
right_frame = Frame(root, bg="red", width=500, height=100)
right_frame.pack(side=LEFT)

photo = Label(left_frame, text="Upload Image")
photo.grid(row=0, column=0, sticky=E)
entry1 = Entry(left_frame)
entry1.grid(row=0, column=1)
entry1.focus()

name = Label(left_frame, text="Enter Your Name")
name.grid(row=1, column=0, sticky=E)
entry2 = Entry(left_frame)
entry2.grid(row=1, column=1)

roll_number = Label(left_frame, text="Enter Your Roll Number")
roll_number.grid(row=2, column=0, sticky=E)
entry3 = Entry(left_frame)
entry3.grid(row=2, column=1)

department = Label(left_frame, text="Enter Your Department")
department.grid(row=3, column=0, sticky=E)
entry4 = Entry(left_frame)
entry4.grid(row=3, column=1)

fa = Label(left_frame, text="Enter Name of FA")
fa.grid(row=4, column=0, sticky=E)
entry5 = Entry(left_frame)
entry5.grid(row=4, column=1)

submit = Button(left_frame, text="Submit")
submit.grid(columns=2)
submit.bind("<Button-1>", (lambda event: check()))

root.mainloop()
```
